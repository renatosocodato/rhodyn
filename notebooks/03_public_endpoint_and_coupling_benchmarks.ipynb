# Public endpoint and bounded-coupling benchmarks

This notebook inspects the Stage 3 public endpoint/model-comparison and bounded-coupling tables. It keeps model comparison and bounded coupling as context-limited software-methods examples.

```python
import csv
from pathlib import Path

root = Path('..') if Path('../case_studies').exists() else Path('.')

def read_rows(path):
    with path.open(newline='', encoding='utf-8') as handle:
        return list(csv.DictReader(handle))
```

The Cell Painting/MitoTox benchmark asks whether reduced endpoint or morphology-only architectures match a public perturbation endpoint table as well as a routed compartment architecture.

```python
ranking = read_rows(root / 'case_studies/cell_painting_mitotox_model_ranking.csv')
[(row['rank'], row['model'], row['weighted_rmse'], row['bic'], row['delta_bic']) for row in ranking]
```

The ERK/Akt benchmark asks whether paired residence summaries remain bounded in the same cells under a declared margin. Passing and failing contexts are both retained so the example does not overstate pathway equivalence.

```python
coupling = read_rows(root / 'case_studies/erk_gpcr_erk_akt_bounded_coupling.csv')
[(row['contrast'], row['estimate'], row['ci_low'], row['ci_high'], row['margin'], row['passes']) for row in coupling]
```

The narrow Stage 3 readout is that RhoDyn can preserve a bounded-coupling decision for one public GPCR context while keeping other contexts outside the claim.

```python
passed = [row['contrast'] for row in coupling if row['passes'] == '1']
failed = [row['contrast'] for row in coupling if row['passes'] == '0']
{'passed': passed, 'failed': failed}
```
