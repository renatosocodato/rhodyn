# Public signaling benchmarks

This notebook inspects the retained Stage 3 public signaling tables. It asks whether amplitude-only and residence-only cells both exist in independent public live-cell systems.

```python
import csv
from collections import Counter
from pathlib import Path

root = Path('..') if Path('../case_studies').exists() else Path('.')

def read_rows(path):
    with path.open(newline='', encoding='utf-8') as handle:
        return list(csv.DictReader(handle))

def class_counts(rows):
    return Counter(row['amplitude_residence_class'] for row in rows)
```

The DRG calcium benchmark is a public neuronal calcium-trace example. The retained table contains episode-cell summaries, not raw imaging files.

```python
drg = read_rows(root / 'case_studies/drg_calcium_residence_amplitude_benchmark.csv')
{'rows': len(drg), 'cells': len({row['cell_id'] for row in drg}), 'classes': dict(class_counts(drg))}
```

The ERK/GPCR benchmark is a public kinase-reporter example. It tests whether peak ERK KTR signal and high-ERK residence identify identical cells.

```python
erk = read_rows(root / 'case_studies/erk_gpcr_residence_amplitude_benchmark.csv')
{
    'rows': len(erk),
    'ligands': dict(Counter(row['ligand'] for row in erk)),
    'classes': dict(class_counts(erk)),
}
```

The Stage 3 gate is supported when both public systems contain amplitude-only and residence-only top-quartile cases. This shows that amplitude summaries and residence summaries are related but not interchangeable.

```python
for label, rows in [('DRG calcium', drg), ('ERK GPCR', erk)]:
    counts = class_counts(rows)
    print(label, 'amplitude_only=', counts['amplitude_only_top_quartile'], 'residence_only=', counts['residence_only_top_quartile'])
```
